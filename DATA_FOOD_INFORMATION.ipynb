```python
!pip install pandas matplotlib seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Datos (por simplicidad, los datos se pueden copiar en formato CSV)
data = {
    'Year': [1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999,
             2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009,
             2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019,
             2020, 2021, 2022],
    'Production (t)': [7066000, 7413000, 4882600, 6501200, 7042900,
                       7270900, 7536400, 7180000, 8952400, 9161300,
                       6953700, 8389200, 8998800, 9587300, 8836800,
                       9332200, 8989800, 11648700, 10592000, 9561200,
                       12043300, 11358700, 13060100, 14190700, 11606400,
                       13679500, 13889000, 14095500, 13884800, 13403900,
                       13563300, 14610792, 14538878],
    'Population': [27657206, 28015762, 28347644, 28668168, 28981190,
                   29289430, 29593856, 29882752, 30144816, 30399776,
                   30683316, 31002768, 31322400, 31619456, 31911670,
                   32215914, 32531864, 32862458, 33218542, 33593920,
                   33963416, 34323536, 34691876, 35063696, 35404616,
                   35732120, 36113532, 36554344, 37035260, 37522584,
                   37888704, 38155012, 38454330]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Calcular la producción per cápita
df['Production per capita (kg)'] = (df['Production (t)'] / df['Population']) * 1000

# Gráfica 1: Producción total a lo largo del tiempo
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Production (t)', marker='o')
plt.title('Producción total en Canadá (1990-2022)')
plt.xlabel('Año')
plt.ylabel('Producción (toneladas)')
plt.grid()
plt.show()

# Gráfica 2: Producción per cápita a lo largo del tiempo
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Production per capita (kg)', marker='o', color='orange')
plt.title('Producción per cápita en Canadá (1990-2022)')
plt.xlabel('Año')
plt.ylabel('Producción per cápita (kg)')
plt.grid()
plt.show()
```
