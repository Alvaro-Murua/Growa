```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data on dengue cases by year in Mexico
data = {
    'ID': ['Mexico'] * 11,
    'Serotype': ['DEN 1,2,3,4'] * 11,
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Epidemiological Week': [53, 52, 52, 52, 52, 52, 53, 52, 52, 52, 38],
    'Total Dengue Cases': [124943, 219593, 130069, 88993, 78621,
                           268458, 120639, 36742, 59918, 277963, 311100],
    'Incidence Rate': [100.92, 175.34, 101.12, 69.6, 60.13,
                       205.31, 92.26, 28.21, 46, 211.81, 235.13],
    'Lab Confirmed Cases': [32100, 26665, 17795, 14138, 12706,
                           43362, 24313, 6746, 12671, 54406, 64815],
    '% Lab Confirmed': [25.69, 12.14, 13.68, 15.73, 16.16,
                       16.15, 20.15, 18.36, 21.15, 19.57, 20.83],
    'Severe Dengue': [8668, 5464, 806, 375, 858,
                      3560, 1058, 300, 410, 1653, 2089],
    'Deaths': [39, 42, 34, 34, 45,
               421, 79, 39, 53, 203, 156],
    'Lethality': [0.031, 0.019, 0.026, 0.038, 0.057,
                  0.157, 0.065, 0.106, 0.088, 0.073, 0.05],
    'Population (per 1000)': [123799, 125236, 128632, 129163, 130759,
                              130759, 130759, 130262, 130262, 131230, 132308]
}

# Create the DataFrame
dengue_data = pd.DataFrame(data)

# Set the visual style
sns.set(style="whitegrid")

# Create a figure with multiple subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Plot Total Dengue Cases over the years
sns.lineplot(x='Year', y='Total Dengue Cases', data=dengue_data, marker='o', ax=axs[0, 0])
axs[0, 0].set_title('Total Dengue Cases in Mexico (2014-2024)')
axs[0, 0].set_ylabel('Total Cases')
axs[0, 0].set_xlabel('Year')

# Plot Incidence Rate over the years
sns.barplot(x='Year', y='Incidence Rate', data=dengue_data, ax=axs[0, 1], palette='Blues')
axs[0, 1].set_title('Incidence Rate of Dengue in Mexico (2014-2024)')
axs[0, 1].set_ylabel('Incidence Rate (per 1000)')
axs[0, 1].set_xlabel('Year')

# Plot Deaths from Dengue over the years
sns.lineplot(x='Year', y='Deaths', data=dengue_data, marker='o', ax=axs[1, 0], color='red')
axs[1, 0].set_title('Deaths due to Dengue in Mexico (2014-2024)')
axs[1, 0].set_ylabel('Total Deaths')
axs[1, 0].set_xlabel('Year')

# Plot Lethality over the years
sns.barplot(x='Year', y='Lethality', data=dengue_data, ax=axs[1, 1], palette='Reds')
axs[1, 1].set_title('Lethality Rate of Dengue in Mexico (2014-2024)')
axs[1, 1].set_ylabel('Lethality Rate (%)')
axs[1, 1].set_xlabel('Year')

# Adjust layout
plt.tight_layout()
plt.show()

```
