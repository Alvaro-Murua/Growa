```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Supply waste data
supply_waste_data = {
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Supply Waste (t)': [3954000, 4325000, 4973000, 5060000, 5197000, 5136000, 1304000, 1225000, np.nan]
}

# Dengue cases data
dengue_data = {
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Total Cases of Dengue': [124943, 219593, 130069, 88993, 78621, 268458, 120639, 36742, 59918]
}

# Create DataFrames
supply_waste_df = pd.DataFrame(supply_waste_data)
dengue_df = pd.DataFrame(dengue_data)

# Merge the DataFrames into one
merged_df = pd.merge(supply_waste_df, dengue_df, on='Year')

# Remove rows with NaN values
merged_df.dropna(inplace=True)

# Calculate the correlation
correlation = merged_df['Supply Waste (t)'].corr(merged_df['Total Cases of Dengue'])
print(f"Correlation between Supply Waste and Dengue Cases: {correlation}")

# Visualize the correlation
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x='Supply Waste (t)', y='Total Cases of Dengue')
plt.title('Correlation between Supply Waste and Dengue Cases')
plt.xlabel('Supply Waste (t)')
plt.ylabel('Total Cases of Dengue')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.show()

```
